```python
# Ensure Ship Date is in datetime format
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

# Extract year into a new column
df['Year'] = df['Ship.Date'].dt.year

# Group by year and aggregate
yearly_summary = df.groupby('Year').agg(
    Total_orders=('Order.ID', 'count'),
    Total_Sales=('Sales', 'sum'),
    Total_Quantity=('Quantity', 'sum'),
    Total_ShippingCost=('Shipping.Cost', 'sum'),
    Total_Discount=('Discount', 'sum'),
    Profit=('Profit', 'sum')
).reset_index()

# Sort by year
yearly_summary = yearly_summary.sort_values(by='Year')

# Optional: round values for better display
yearly_summary = yearly_summary.round(2)

# Display
print('                       Yearly Orders, Sales & Profit\n')
print(yearly_summary.to_string(index=False))
```

```python
# Convert Ship.Date to datetime
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

# Extract month
df['Month'] = df['Ship.Date'].dt.month

# Define season conditions
conditions = [
    (df['Month'].between(1, 3)),
    (df['Month'].between(4, 6)),
    (df['Month'].between(7, 9)),
    (df['Month'].between(10, 12))
]

choices = ['Jan-Mar', 'Apr-Jun', 'Jul-Sep', 'Oct-Dec']

# Fix: Add default value of the same type (string)
df['Season'] = np.select(conditions, choices, default='Unknown')

# Group by Season
season_summary = df.groupby('Season').agg(
    Total_orders=('Order.ID', 'count'),
    Total_Sales=('Sales', 'sum'),
    Total_Quantity=('Quantity', 'sum'),
    Total_ShippingCost=('Shipping.Cost', 'sum'),
    Total_Discount=('Discount', 'sum'),
    Profit=('Profit', 'sum')
).reset_index()

# Sort Season in the same order as SQL's FIELD()
season_order = ['Jan-Mar', 'Apr-Jun', 'Jul-Sep', 'Oct-Dec']
season_summary['Season'] = pd.Categorical(season_summary['Season'], categories=season_order, ordered=True)
season_summary = season_summary.sort_values('Season')

# Optional: round values
season_summary = season_summary.round(2)

# Display
print('                     Season-wise Orders, Sales & Profit\n')
print(season_summary.to_string(index=False))
```

```python
# Group by Category and aggregate sales and profit
category_summary = df.groupby('Category').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

# Calculate Profit Margin %
category_summary['Profit_Margin_Percentage'] = (
    (category_summary['Total_Profit'] / category_summary['Total_Sales']) * 100
).round(2)

# Sort by Profit Margin %
category_summary = category_summary.sort_values(by='Profit_Margin_Percentage', ascending=False)

# Display
print('                     Profit Margins by Category\n')
print(category_summary.to_string(index=False))
```

```python
# Create shipping cost buckets
df['ShippingCost_bucket'] = pd.cut(
    df['Shipping.Cost'],
    bins=[-float('inf'), 20, 50, float('inf')],
    labels=['Low (<20)', 'Medium (20-50)', 'High (>50)']
)

# Group by bucket and calculate metrics
shipping_summary = df.groupby('ShippingCost_bucket').agg(
    orders=('Order.ID', 'count'),
    total_sales=('Sales', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

# Calculate profit margin
shipping_summary['profit_margin'] = (
    (shipping_summary['total_profit'] / shipping_summary['total_sales']) * 100
).round(2)

# Sort by bucket order (already sorted if using pd.cut labels)
print("               Profit vs Shipping Cost Buckets\n")
print(shipping_summary.to_string(index=False))
```

```python
discount_profit = df.groupby('Discount', as_index=False).agg(
    Total_profit=('Profit', 'sum')
).sort_values(by='Discount')

print("             Total Profit by Discount Level\n")
print(discount_profit.to_string(index=False))
```

```python
# Group and aggregate profit by discount
discount_profit = df.groupby('Discount', as_index=False).agg(
    Total_profit=('Profit', 'sum')
).sort_values(by='Discount')

# Line chart
mpl.figure(figsize=(12, 5))
mpl.plot(discount_profit['Discount'], discount_profit['Total_profit'], marker='o', color='teal')

mpl.title('Total Profit vs Discount Levels', fontsize=14)
mpl.xlabel('Discount', fontsize=12)
mpl.ylabel('Total Profit', fontsize=12)
mpl.grid(True)

mpl.tight_layout()

mpl.show()
```

```python
# Create shipping cost buckets
df['ShippingCost_bucket'] = pd.cut(
    df['Shipping.Cost'],
    bins=[-float('inf'), 20, 50, float('inf')],
    labels=['Low (<20)', 'Medium (20-50)', 'High (>50)']
)

# Group by bucket and calculate metrics
shipping_summary = df.groupby('ShippingCost_bucket').agg(
    orders=('Order.ID', 'count'),
    total_sales=('Sales', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

# Calculate profit margin
shipping_summary['profit_margin'] = (
    (shipping_summary['total_profit'] / shipping_summary['total_sales']) * 100
).round(2)

# Sort by bucket order (already sorted if using pd.cut labels)
print("               Profit vs Shipping Cost Buckets\n")
print(shipping_summary.to_string(index=False))
```

```python
# Set style
sb.set(style='whitegrid')

# Create the bar plot
mpl.figure(figsize=(8, 5))
sb.barplot(data=shipping_summary, x='ShippingCost_bucket', y='profit_margin', palette='viridis')

# Titles and labels
mpl.title('Profit Margin by Shipping Cost Bucket', fontsize=14)
mpl.xlabel('Shipping Cost Bucket', fontsize=12)
mpl.ylabel('Profit Margin (%)', fontsize=12)

# Show value labels on bars
for i, val in enumerate(shipping_summary['profit_margin']):
    mpl.text(i, val + 0.5, f'{val}%', ha='center', fontsize=10)

mpl.tight_layout()
mpl.show()
```
